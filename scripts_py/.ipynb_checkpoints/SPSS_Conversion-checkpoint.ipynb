```python
for i in range(1,19):
    folder = f"bhps_w{i}"
    for file in os.listdir(folder):
        if file.endswith(".sav"):
            path = folder + "\\" + file
            csv = file.split(".")[0]
            export_path = folder + "\\" + csv + ".csv"
            df = pd.read_spss(path)
            df.to_csv(export_path)
```

### Calculation of income inequality by Gini index in each county

```python
import numpy as np
import pandas as pd

# convert "afiyr" column to float data type
df_residence["afiyr"] = df_residence["afiyr"].astype("float32")

# group incomes by county
grouped = df_residence.groupby("aplb4d")

# define a function to calculate the Gini index
def gini_coefficient(x):
    # sort incomes in ascending order
    x = np.sort(x)
    # calculate cumulative share of income
    cum_income = np.cumsum(x) / np.sum(x)
    # calculate cumulative share of population
    cum_pop = np.arange(1, len(x)+1) / len(x)
    # calculate area between Lorenz curve and line of perfect equality
    area = np.trapz(cum_income, cum_pop)
    # calculate Gini coefficient
    gini = 2*area - 1
    return gini

# apply the function to each county group
gini_by_county = grouped["afiyr"].apply(gini_coefficient)

# create a new DataFrame with the results
gini_df = pd.DataFrame({"Gini Index": gini_by_county})

# display the results
print(gini_df)
```

### Calculation of variability of the highest educational qualifcation in each county

```python
import numpy as np
import pandas as pd

# group individuals by county
grouped = df_residence.groupby("aplb4d")

# define a function to calculate the dissimilarity index
def dissimilarity_index(x):
    # calculate the proportion of individuals in each occupation category
    p = x["afetype"].value_counts(normalize=True)
    # calculate the proportion of individuals in each occupation category for the complement group
    q = 1 - p
    # calculate the dissimilarity index
    dissimilarity = 0.5 * np.sum(np.abs(p - q))
    return dissimilarity

# apply the function to each county group
dissimilarity_by_county = grouped.apply(dissimilarity_index)

# create a new DataFrame with the results
dissimilarity_df = pd.DataFrame({"Dissimilarity Index": dissimilarity_by_county})

# display the results
print(dissimilarity_df)
```

```python
def duca(x):
    county_tuple = x["dkp"]
    freq = list(county_tuple[1].value_counts())
    val = list(county_tuple[1].unique())
    norm_freq = [i/sum(freq) for i in freq]
    outer_sum = 0
    for i in range(len(norm_freq)):
        nkit = norm_freq[i]
        inner_sum = 0
        for j in range(len(norm_freq)):
            npit = norm_freq[j]
            dkp = abs(val[i] - val[j])
            inner_sum += npit*dkp
    outer_sum += (nkit**2)*inner_sum
    
    return outer_sum
```

### Calculation of Job status variability using dissimilarity index

```python
def dissimilarity_index(x):
    job = x["job"][1]
    freq = list(job.value_counts())
    s = len(freq)
    dissim_index = 0
    for i in freq:
        dissim_index = 0.5 * abs(i-(1/s))
    return dissim_index

def county_name(x):
    return x["job"][0]

def frac_employed(x):
    job = list(x["job"][1])
    frac_emp = (job.count(2)/len(job))
    return frac_emp
```

```python
job_df = pd.DataFrame()

ajbstatt_dict = {"Self employed" : 1, "In paid employ" : 2, "Unemployed" : 3, "Retired" : 4, "Family care" : 5, \
                    "FT student" : 6,  "Long term sick/disabled" : 7, "On matern leave" : 8, \
                    "Govt trng scheme" : 9, "Something else" : 10, "don't know" : -1, "missing or wild" : -9, \
                    "inapplicable" : -8, "proxy" : -7, "refused" : -2}


# Create a new column "aopsocc_encoded" in the dataframe by replacing the values with encoded values from the dictionary
df_residence["ajbstatt_encoded"] = df_residence["ajbstatt"].replace(ajbstatt_dict)

# Group the dataframe by county and calculate the mean and standard deviation of the encoded values
job_df["job"] = df_residence.groupby("aplb4d")["ajbstatt_encoded"]

job_df["job_variability"] = job_df.apply(lambda x: dissimilarity_index(x), axis = 1)
job_df["county"] = job_df.apply(lambda x: county_name(x), axis = 1)
job_df ["frac_employed"] = job_df.apply(lambda x: frac_employed(x), axis = 1)

job_df
```

### Calculation of variability of educational qualification using dissimilarity index

### Calculation of share of natives in each county

```python
def native_share(x):
    pop_count = list(x["born"][1])
    native_count = pop_count.count("inapplicable")/len(pop_count)
    return native_count
```

```python
native_df = pd.DataFrame()

# Group the dataframe by county and calculate the mean and standard deviation of the encoded values
native_df["born"] = df_residence.groupby("aplb4d")["aplbornc_cc"]

native_df["native_share"] = native_df.apply(lambda x: native_share(x), axis = 1)

native_df
```
